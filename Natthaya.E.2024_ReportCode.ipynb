```python
import pandas as pd
import os 
import glob
```

Import all data into a pandas data frame using pandas, os and glob

```python
path = 'Data' 
all_files = sorted(glob.glob(os.path.join(path, '*.txt'))) #sorted by name 
data_list = []

for eachfile in all_files: 
    data = open(eachfile, "r")
    for line in data.readlines(): #each list item is each line in file
        linebreak = line.split('\t') #removes \
        data_list.append(linebreak) #append next file as list item 
        
second_list = data_list [:2] #takes first two items of data_list (header + data row)
for num in range (3, 98, 2): #start from 4th item (data of file 2), go up by increments of 2(skip header of every file)
    second_list.append(data_list[num])

all_df = pd.DataFrame(second_list[1:], columns = second_list[0])
all_df
```

Isolating just the sequences and gene name. Not included in the report, just for easier indexing if needed later on. 

```python
sequence_df = all_df.iloc[::, 0:3]
```

Determining what type of mutation the gene has (insertion, deletion, substitution) based on the difference in sequence length from the WT sequence.

```python
x = 0 
name_list = []
mutation_list = []
difference_list = []
for x in range (len(sequence_df)): 
    WT = sequence_df.iloc[x, 1]
    M = sequence_df.iloc[x, 2]
    name = sequence_df.iloc[x,0]
    if len(WT) < len(M): # WT < M therefore insertion 
        difference1 = len(M)-len(WT)
        name_list.append(name)
        mutation_list.append('Insertion')
        difference_list.append(difference1)
    elif len(WT) > len(M): # WT > M therefore deletion 
        difference2 = len(WT)-len(M)
        name_list.append(name)
        mutation_list.append('Deletion')
        difference_list.append(difference2)
    elif len(WT) == len(M): # equal length 
        name_list.append(name)
        mutation_list.append('Substitution')
        difference_list.append(0) #no base difference 
    x += 1

mutation_dict = {
    'Gene': name_list, 'Mutation': mutation_list, 'Hamming': difference_list}

mutation_df = pd.DataFrame.from_dict(mutation_dict)
mutation_df
```

Generating a data frame for counting the numbers of each mutation. This is saved as data frame because it will be used in a plot later. 

```python
mutation_counts = mutation_df.groupby('Mutation').size()
#group by column 'Mutation' and count occurences
mutation = ['Deletion', 'Insertion', 'Substitution']
count = [14, 13, 22]

mut_series = pd.Series(mutation)
c_series = pd.Series(count)

framemut = {'Mutation': mut_series, 
           'Count': c_series}
mut_count = pd.DataFrame(framemut)
mut_count
```

The objects in the data frame must be converted to floats in order to perform any further mathematical analyses

```python
pd.options.mode.copy_on_write = True #to prevent SettingWithCopyWarning 

mRNA_WT = all_df[['Gene','mRNA.Expression.WT.Rep1', 'mRNA.Expression.WT.Rep2', 'mRNA.Expression.WT.Rep3']]
mRNA_WT.dtypes #currently data type is objects, need floats 
mRNA_WT[['mRNA.Expression.WT.Rep1', 'mRNA.Expression.WT.Rep2', 'mRNA.Expression.WT.Rep3']] = mRNA_WT[['mRNA.Expression.WT.Rep1', 'mRNA.Expression.WT.Rep2', 'mRNA.Expression.WT.Rep3']].astype(float)
print(mRNA_WT.dtypes) #check datatype 
```

```python
mRNA_WT['Mean.WT.mRNA'] = mRNA_WT.iloc[:, 1:4].mean(axis=1) 
#this takes mean value from 2nd-4th column, axis 1 means columns
mRNA_WT
```

Combining all mean values togehter into 1 data frame in order to find the difference between each mean value. 

```python
WTmRNA = meanmRNA_WT['Mean.WT.mRNA']
MmRNA = meanmRNA_M['Mean.Mut.mRNA']
WTp = meanprotein_WT['Mean.WT.protein']
Mp = meanprotein_M['Mean.M.protein']
WTcv = meanCV_WT['Mean.WT.Viability']
Mcv = meanCV_M['Mean.M.Viability']
mut_type = mutation_df[['Gene','Mutation']]

frames = [mut_type, WTmRNA, MmRNA, WTp, Mp, WTcv, Mcv]
mean_values = pd.concat(frames, axis=1) #axis = 1 = columns
mean_values
```

The difference in mean value (overall change afte the mutation) is calculated by (WT - M). Therefore, a negative difference would mean that the mutant strain had increased expression or viability. 

Add the 3 columns containing the difference between mean mRNA/protein/cell viability of WT versus mutant strain into the main table

```python
mean_values['mRNA.Difference'] = mean_values['Mean.WT.mRNA'].sub(mean_values['Mean.Mut.mRNA'])
mean_values['protein.Difference'] = mean_values['Mean.WT.protein'].sub(mean_values['Mean.M.protein'])
mean_values['CellViability.Difference'] = mean_values['Mean.WT.Viability'].sub(mean_values['Mean.M.Viability'])
mean_values
```

Determining whether the mutation has caused the mRNA expression, protein expression, and cell viability to increase, decrease, or no difference. 

This either returns a False or True statement, which have been saved as a dataframe for future use. 

```python
lessormore_mRNA = mean_values['Mean.WT.mRNA'] > mean_values['Mean.Mut.mRNA']
lessormore_protein = mean_values['Mean.WT.protein'] > mean_values['Mean.M.protein']
lessormore_CV = mean_values['Mean.WT.Viability'] > mean_values['Mean.M.Viability']
mut_type = mutation_df[['Gene','Mutation']]
lessormore_frame = [mut_type, lessormore_mRNA, lessormore_protein, lessormore_CV]


lessormore_data = pd.concat(lessormore_frame, axis=1)
lessormore_data.columns = ['Gene', 'Mutation','WT > M mRNA', 'WT > M protein', 'WT > M CV']
lessormore_data
```

```python
#finding length of each WT sequence 
x = 0 
length_list = []
for x in range (len(all_df['WildType.Sequence'])): 
    WTseq = all_df.iloc[x, 1]
    length_list.append(len(WTseq))
    x += 1

WTlength_df = pd.DataFrame(length_list)
```

Prepare a table with WT gene length and mutant gene length to find any mutation within the first 1000 sequences of the WT sequence 

```python
geneseq = all_df.iloc[:,0:1]
mutation_frame = [geneseq, WTlength_df]

mutation_location = pd.concat(mutation_frame, axis =1)
mutation_location.columns = ['Gene', 'WT length']
```

```python
# finding out whether mutation is in the promoter of cds 
x = 0 
Mutation_Finder = []

for x in range (len(all_df.iloc[::])):
    WT_promoter = (all_df.iloc[x,1])[:1000]
    M_seq = all_df.iloc[x,2]
    if len(M_seq) <= 1000: # if the mutant sequence is < WT promoter region
        Mutation_Finder.append('Promoter') # then mutation must be in promoter
    elif (M_seq)[:1000] != WT_promoter: 
        Mutation_Finder.append('Promoter')
    elif (M_seq)[:1000] == WT_promoter: 
        Mutation_Finder.append('Coding Sequence')
    x += 1 
    
MutLoc_df = pd.DataFrame(Mutation_Finder)
```

```python
# saved as data frame to add M seq length to mutation_location
x = 0 
length_list2 = []
for x in range (len(all_df['Mutant.Sequence'])): 
    Mseq = all_df.iloc[x, 2]
    length_list2.append(len(Mseq))
    x += 1

Mlength_df = pd.DataFrame(length_list2)
```

```python
MutLoc_frames = [mutation_location, Mlength_df, MutLoc_df]

Mut_Where = pd.concat(MutLoc_frames, axis = 1)
Mut_Where.columns = ['Gene', 'WT Length', 'M Length', 'Mut Location']
Mut_Where
```
